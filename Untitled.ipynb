```python
import graphviz

# Crear un diagrama de prueba muy básico
dot = graphviz.Digraph('Prueba')
dot.node('A', '¡Graphviz Funciona!')

# Intentar renderizar y mostrar
try:
    dot.render('prueba_instalacion', format='png', view=True)
    print("¡Instalación exitosa! El motor fue detectado correctamente.")
    display(dot)
except Exception as e:
    print(f"La instalación aún no es detectada. Error: {e}")
```
